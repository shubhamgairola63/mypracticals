```python
!pip install osmnx
# Practical 6A
import osmnx as ox

# Load and project the graph
place_name = "Mumbai, Maharashtra, India" # Made the place name more specific
G = ox.graph_from_place(place_name, network_type="drive")
G_proj = ox.project_graph(G)

# Convert to GeoDataFrames
nodes, edges = ox.graph_to_gdfs(G_proj)

# Use .explore() for an interactive OSM map
m = edges.explore(
    color="black",
    style_kwds={'weight': 1, 'opacity': 0.7},
    tiles="OpenStreetMap",
    name="Road Network"
)

# Display the map
m
```

```python
# Practical 6B
import osmnx as ox
import folium

# Define the center point
address = "Marine Drive, Mumbai, India"
dist = 800

# Download the networks
G_drive = ox.graph_from_address(address, dist=dist, network_type="drive")
G_walk = ox.graph_from_address(address, dist=dist, network_type="walk")

# Convert Graphs to GeoDataFrames
nodes_d, edges_drive = ox.graph_to_gdfs(G_drive)
nodes_w, edges_walk = ox.graph_to_gdfs(G_walk)

# Plot the walking network (Blue)
m = edges_walk.explore(
    color="blue",
    style_kwds={'weight': 2, 'opacity': 0.6},
    name="Walking Network"
)

# Overlay the driving network (Red)
edges_drive.explore(
    m=m,
    color="red",
    style_kwds={'weight': 4, 'opacity': 0.7},
    name="Driving Network"
)

# Add Layer Control
folium.LayerControl().add_to(m)
m
```
